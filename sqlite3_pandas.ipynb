run basic queries using pandas

the data will be populated is in a `csv` without headers but the headers are `id`, `department_name`, `manager_id`, `location_id`

```python
%%bash
wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMSkillsNetwork-PY0221EN-Coursera/labs/v2/Departments.csv
```

```python
import pandas as pd
import sqlite3
```

```python
db_name = "staff"
table_name = "department"
```

```python
conn = sqlite3.connect(db_name)
```

```python
columns = ["id", "name", "manager_id", "location_id"]
```

```python
df = pd.read_csv("Departments.csv", names=columns)
df.head()
```

```python
df.to_sql(table_name, conn, if_exists="replace", index=False)
```

```python
query_1 = f"select * from {table_name}"
query_2 = f"select name from {table_name}"
query_3 = f"select count(*) from {table_name}"
```

```python
output = pd.read_sql(query_1, conn)
print(query_1)
output
```

```python
output = pd.read_sql(query_2, conn)
print(query_2)
output
```

```python
output = pd.read_sql(query_3, conn)
print(query_3)
output
```

```python
conn.close()
```
