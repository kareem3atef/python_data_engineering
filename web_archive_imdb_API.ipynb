```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```python
url = "https://web.archive.org/web/20240103020246/https://www.imdb.com/chart/top/"
```

```python
page = requests.get(url).text
data = BeautifulSoup(page, "html.parser")
```

```python
tables = data.find_all("ul")
table = tables[12]
```

```python
films = table.find_all("li")
```

```python
df = pd.DataFrame(columns=["film", "rating", "year", "img"])
```

```python
for film in films:
    film_name = film.find("h3").text
    film_name = re.split(r"\d+\.\s", film_name)[1]
    rating = film.find("svg", class_="ipc-icon--star-inline").next_sibling
    year = film.find("span", class_="sc-43986a27-8 jHYIIK cli-title-metadata-item").text
    img = film.find("img", class_="ipc-image").get("src")
    film_dict = {
        "film": film_name,
        "rating": rating,
        "year": year,
        "img": img
    }
    df1 = pd.DataFrame(film_dict, index=[0])
    df = pd.concat([df, df1], ignore_index=True)
```

```python
pd.set_option("max_colwidth", True)
```

```python
df.head()
```

```python
df.to_csv("imdb_top_films.csv", index=False)
```
